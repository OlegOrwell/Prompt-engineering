## The task is to extract all questions manager asked about clients health.

### Define function to build prompts:

```python
def build_prompt(text):
    prompt_left = """Ты умный поисковик строчка за строчкой просмотри диалог. Твоя задача найти все вопросы менеджера касающиеся здоровья клиента. Если не уверен в ответе - пиши "Неизвестно". Диалог в тройных кавычках: '''"""
    prompt_right = """'''. Выведи найденные вопросы списком;"""
    return prompt_left + text + prompt_right
```

### Define function to make requests:

```python
def insight(prompt, text='', tag='', temperature=0): 
    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      temperature = temperature,
      messages=[
          {"role": "system", "content": f"""{prompt} {text} {tag}"""},
      ]
    )
    return response.choices[0].message.content
```

#### Argument "tag" is used in case I need to pass to the model more specific instructions of how to form output.

```python
prompt_template = f"""Ты умный поисковик строчка за строчкой просмотри диалог. Твоя задача найти все вопросы менеджера касающиеся здоровья клиента. Если не уверен в ответе - пиши "Неизвестно". Диалог в тройных кавычках: '''{text_1}'''. Выведи найденные вопросы списком;"""
```

```python
def build_prompt_(text):
    prompt_left = """Ты умный поисковик строчка за строчкой просмотри диалог. Твоя задача найти все реплики клиента в которых он признаёт наличие у себя болезней. Если не уверен в ответе - пиши "Неизвестно". Диалог в тройных кавычках: '''"""
    prompt_right = """'''. Выведи ответы клиента списком;"""
    return prompt_left + text + prompt_right
```
